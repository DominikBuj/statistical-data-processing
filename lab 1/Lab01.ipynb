#### Zadanie 3

Wyświetl tylko dane z roku `2008`.


````python
import pandas as pd

def getDataFrameFromFile(filePath):
    with open(filePath, 'rb') as file:
        content = np.load(file)
        columns, data = content['columns'], content['data']
        dataFrame = pd.DataFrame()
        for index, column in enumerate(columns):
            dataFrame[column] = data[:, index]
        return dataFrame

expensesDataFrame = getDataFrameFromFile('./dataset/dane.npz')

display(expensesDataFrame[expensesDataFrame['Rok'] == '2008'])
````

#### Zadanie 4

Wyświetl wykres uwzględniając 10 województw (bez sumarycznego dla całego kraju), w których wydatki budżetów były największe.

> Podpowiedź: Możesz w tym celu wykorzystać moduł matplotlib.pyplot. Przykład:
```python
import matplotlib.pyplot as plt
plt.bar([1, 2, 3], [5, 2, 4])
```

> Podpowiedź 2: Sprawdź inne funkcje aby utworzyć tytuł wykresu, oznaczyć osie, wyświetlić legendę i zmienić kolory słupków. https://matplotlib.org/stable/tutorials/introductory/pyplot.html

````python
import matplotlib.pyplot as plt

topExpensesDataFrame = pd.DataFrame({'Województwo':expensesDataFrame['Nazwa'], 'Wydatki':expensesDataFrame['Wartosc'].astype(float)})
topExpensesDataFrame = topExpensesDataFrame[topExpensesDataFrame['Województwo'] != 'POLSKA']
topExpensesDataFrame = topExpensesDataFrame.groupby(['Województwo'], as_index=False).sum()
topExpensesDataFrame = topExpensesDataFrame.nlargest(10, 'Wydatki')[::-1]

plt.figure(figsize=(24, 8))
plt.title('Top 10 polskich województw z największymi wydatkami')
plt.xlabel('Województwo')
plt.ylabel('Wydatki [zł]')
plt.bar(topExpensesDataFrame['Województwo'], topExpensesDataFrame['Wydatki'])
plt.show()
````

#### Zadanie 5

Wczytaj dane z pliku `dane-ludnosc.npz`. Plik ten zawiera zbiór danych zawierający **liczbę ludnosci**.

Oblicz wydatki z budżetu na mieszkańca dla każdego z województw (_uwzględnij dane dla roku `2008`_).


````python
populationDataFrame = getDataFrameFromFile('./dataset/dane-ludnosc.npz')

expenses2008DataFrame = expensesDataFrame[expensesDataFrame['Rok'] == '2008']
expenses2008DataFrame = expenses2008DataFrame[expenses2008DataFrame['Nazwa'] != 'POLSKA']
expenses2008DataFrame = expenses2008DataFrame[['Nazwa', 'Wartosc']]

population2008DataFrame = populationDataFrame[expensesDataFrame['Rok'] == '2008']
population2008DataFrame = population2008DataFrame[population2008DataFrame['Nazwa'] != 'POLSKA']
population2008DataFrame = population2008DataFrame[['Nazwa', 'Wartosc']]

personExpenses2008DataFrame = pd.DataFrame()
personExpenses2008DataFrame['Województwo'] = expenses2008DataFrame['Nazwa']
personExpenses2008DataFrame['Wydatki'] = expenses2008DataFrame['Wartosc'].astype(float) / population2008DataFrame['Wartosc'].astype(float)
personExpenses2008DataFrame = personExpenses2008DataFrame.sort_values(by=['Wydatki'])

plt.figure(figsize=(40, 8))
plt.title('Wydatki z budżetu na mieszkańca dla każdego z województw z roku 2008')
plt.xlabel('Województwo')
plt.ylabel('Wydatki [zł]')
plt.bar(personExpenses2008DataFrame['Województwo'], personExpenses2008DataFrame['Wydatki'])
plt.show()
````

#### Zadanie 6

Wykonaj dalszą analizę zbiorów danych (dowolną). Czy coś ciekawego zaobserwowałeś?


````python
# Zmiana wydatków budżetowych dla województw polskich przez lata 1999-2020

firstExpensesDataFrame = expensesDataFrame[expensesDataFrame['Rok'] == '1999'][['Nazwa', 'Wartosc']]
lastExpensesDataFrame = expensesDataFrame[expensesDataFrame['Rok'] == '2020'][['Nazwa', 'Wartosc']]

allExpensesDataFrame = firstExpensesDataFrame.merge(lastExpensesDataFrame, on=['Nazwa'])
allExpensesDataFrame = allExpensesDataFrame.rename(columns={'Nazwa':'Województwo', 'Wartosc_x':'1999', 'Wartosc_y':'2020'})
allExpensesDataFrame[['1999', '2020']] = allExpensesDataFrame[['1999', '2020']].astype(float)
allExpensesDataFrame['Zmiana'] = (allExpensesDataFrame['2020'] - allExpensesDataFrame['1999']) / allExpensesDataFrame['2020'] * 100.0
allExpensesDataFrame = allExpensesDataFrame.sort_values(by=['Zmiana'])

plt.figure(figsize=(40, 8))
plt.title('Zmiana wydatków budżetowych dla województw polskich przez lata 1999-2020')
plt.xlabel('Województwo')
plt.ylabel('Zmiana wydatków budżetowych [%]')
plt.bar(allExpensesDataFrame['Województwo'], allExpensesDataFrame['Zmiana'])
plt.show()

# Wzrost wydatków budżetowych dla województw polskich jest porównywalny dla wszystkich województw
# Ilość wydatków budżetowych dla województwa nie jest proporcjonalna do wzrostu tych wydatków
````
