#### Zadanie 3

Wywietl tylko dane z roku `2008`.


````python
import pandas as pd

def getDataFrameFromFile(filePath):
    with open(filePath, 'rb') as file:
        content = np.load(file)
        columns, data = content['columns'], content['data']
        dataFrame = pd.DataFrame()
        for index, column in enumerate(columns):
            dataFrame[column] = data[:, index]
        return dataFrame

expensesDataFrame = getDataFrameFromFile('./dataset/dane.npz')

display(expensesDataFrame[expensesDataFrame['Rok'] == '2008'])
````

#### Zadanie 4

Wywietl wykres uwzgldniajc 10 wojew贸dztw (bez sumarycznego dla caego kraju), w kt贸rych wydatki bud偶et贸w byy najwiksze.

> Podpowied藕: Mo偶esz w tym celu wykorzysta modu matplotlib.pyplot. Przykad:
```python

```

> Podpowied藕 2: Sprawd藕 inne funkcje aby utworzy tytu wykresu, oznaczy osie, wywietli legend i zmieni kolory supk贸w. https://matplotlib.org/stable/tutorials/introductory/pyplot.html

````python
import matplotlib.pyplot as plt

topExpensesDataFrame = pd.DataFrame({'Wojew贸dztwo':expensesDataFrame['Nazwa'], 'Wydatki':expensesDataFrame['Wartosc'].astype(float)})
topExpensesDataFrame = topExpensesDataFrame[topExpensesDataFrame['Wojew贸dztwo'] != 'POLSKA']
topExpensesDataFrame = topExpensesDataFrame.groupby(['Wojew贸dztwo'], as_index=False).sum()
topExpensesDataFrame = topExpensesDataFrame.nlargest(10, 'Wydatki')[::-1]

plt.figure(figsize=(24, 8))
plt.title('Top 10 polskich wojew贸dztw z najwikszymi wydatkami')
plt.xlabel('Wojew贸dztwo')
plt.ylabel('Wydatki [z]')
plt.bar(topExpensesDataFrame['Wojew贸dztwo'], topExpensesDataFrame['Wydatki'])
plt.show()
````

#### Zadanie 5

Wczytaj dane z pliku `dane-ludnosc.npz`. Plik ten zawiera zbi贸r danych zawierajcy **liczb ludnosci**.

Oblicz wydatki z bud偶etu na mieszkaca dla ka偶dego z wojew贸dztw (_uwzgldnij dane dla roku `2008`_).


````python
populationDataFrame = getDataFrameFromFile('./dataset/dane-ludnosc.npz')

expenses2008DataFrame = expensesDataFrame[expensesDataFrame['Rok'] == '2008']
expenses2008DataFrame = expenses2008DataFrame[expenses2008DataFrame['Nazwa'] != 'POLSKA']
expenses2008DataFrame = expenses2008DataFrame[['Nazwa', 'Wartosc']]

population2008DataFrame = populationDataFrame[expensesDataFrame['Rok'] == '2008']
population2008DataFrame = population2008DataFrame[population2008DataFrame['Nazwa'] != 'POLSKA']
population2008DataFrame = population2008DataFrame[['Nazwa', 'Wartosc']]

personExpenses2008DataFrame = pd.DataFrame()
personExpenses2008DataFrame['Wojew贸dztwo'] = expenses2008DataFrame['Nazwa']
personExpenses2008DataFrame['Wydatki'] = expenses2008DataFrame['Wartosc'].astype(float) / population2008DataFrame['Wartosc'].astype(float)
personExpenses2008DataFrame = personExpenses2008DataFrame.sort_values(by=['Wydatki'])

plt.figure(figsize=(40, 8))
plt.title('Wydatki z bud偶etu na mieszkaca dla ka偶dego z wojew贸dztw z roku 2008')
plt.xlabel('Wojew贸dztwo')
plt.ylabel('Wydatki [z]')
plt.bar(personExpenses2008DataFrame['Wojew贸dztwo'], personExpenses2008DataFrame['Wydatki'])
plt.show()
````

#### Zadanie 6

Wykonaj dalsz analiz zbior贸w danych (dowoln). Czy co ciekawego zaobserwowae?


````python
# Zmiana wydatk贸w bud偶etowych dla wojew贸dztw polskich przez lata 1999-2020

firstExpensesDataFrame = expensesDataFrame[expensesDataFrame['Rok'] == '1999'][['Nazwa', 'Wartosc']]
lastExpensesDataFrame = expensesDataFrame[expensesDataFrame['Rok'] == '2020'][['Nazwa', 'Wartosc']]

allExpensesDataFrame = firstExpensesDataFrame.merge(lastExpensesDataFrame, on=['Nazwa'])
allExpensesDataFrame = allExpensesDataFrame.rename(columns={'Nazwa':'Wojew贸dztwo', 'Wartosc_x':'1999', 'Wartosc_y':'2020'})
allExpensesDataFrame[['1999', '2020']] = allExpensesDataFrame[['1999', '2020']].astype(float)
allExpensesDataFrame['Zmiana'] = (allExpensesDataFrame['2020'] - allExpensesDataFrame['1999']) / allExpensesDataFrame['2020'] * 100.0
allExpensesDataFrame = allExpensesDataFrame.sort_values(by=['Zmiana'])

plt.figure(figsize=(40, 8))
plt.title('Zmiana wydatk贸w bud偶etowych dla wojew贸dztw polskich przez lata 1999-2020')
plt.xlabel('Wojew贸dztwo')
plt.ylabel('Zmiana wydatk贸w bud偶etowych [%]')
plt.bar(allExpensesDataFrame['Wojew贸dztwo'], allExpensesDataFrame['Zmiana'])
plt.show()

# Wzrost wydatk贸w bud偶etowych dla wojew贸dztw polskich jest por贸wnywalny dla wszystkich wojew贸dztw
# Ilo wydatk贸w bud偶etowych dla wojew贸dztwa nie jest proporcjonalna do wzrostu tych wydatk贸w
````
